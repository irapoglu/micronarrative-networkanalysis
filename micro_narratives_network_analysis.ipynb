```python
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd

# Veri setini yükleme
df = pd.read_excel("/content/data.xlsx")

# Sadece birinci katılımcının verilerini filtreleme
df_filtered = df[df['participants'] == 'participant 1']  # 'participant1' doğru ismi kullandığınızdan emin olun

# Temalara göre renkler tanımlama
theme_colors = {
    'social environment': 'green',
    'physical environment': 'blue',
    'environmental meaning': 'red'
}

# Ağ grafiği oluşturma
G = nx.Graph()

# Dönem düğümleri
for period in df_filtered['periods'].unique():
    G.add_node(period, color='lightblue')  # Dönem düğümlerine mavi renk

# Tema ve kavram düğümleri ekleme
for index, row in df_filtered.iterrows():
    theme_color = theme_colors.get(row['theme'], 'gray')
    G.add_node(row['concepts'], color=theme_color)  # Kavram düğümlerine tema rengi ekleme
    G.add_edge(row['periods'], row['concepts'], color='gray')  # Kenarlar gri renk

# Spring layout ile pozisyon belirleme (daha geniş k ve scale değeri ile)
pos = nx.spring_layout(G, k=0.5, iterations=50, seed=42)  # k değeri, düğümler arasındaki minimum mesafeyi kontrol eder

# Düğüm derecelerini hesaplama (düğüm büyüklükleri için)
degree_dict = dict(G.degree(G.nodes()))
sizes = [v * 300 for v in degree_dict.values()]  # Düğüm boyutlarını derecelerle orantılı olarak ayarlama

# Düğüm renkleri
node_colors = [nx.get_node_attributes(G, 'color')[node] for node in G]

# Görselleştirme
plt.figure(figsize=(12, 6))
nx.draw(G, pos, node_color=node_colors, node_size=sizes, edge_color='gray', alpha=0.6)  # Düğümlerin opaklığı

# Küçük harfe çevrilen etiketleri çizme
labels = {node: node.lower() for node in G.nodes()}
nx.draw_networkx_labels(G, pos, labels=labels, font_size=10, font_color='black', alpha=0.9)  # Yazıların opaklığı


plt.title('Network Graph for Participant 1 with Node Sizes by Degree')
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd

# Veri setini yükleme
df = pd.read_excel("/content/data.xlsx")

# Temalara göre renkler tanımlama
theme_colors = {
    'social environment': 'green',
    'physical environment': 'blue',
    'environmental meaning': 'red'
}

# Katılımcı renklerini tanımlama
participant_colors = {
    'participant 1': 'purple',
    'participant 2': 'orange',
    'participant 3': 'cyan',
    'participant 4': 'pink',
    # Diğer katılımcılar için renkler ekleyin
}

# Ağ grafiği oluşturma
G = nx.Graph()

# Tüm katılımcıları içeren ağ grafiği oluşturma
for index, row in df.iterrows():
    participant = row['participants']
    period = row['periods']
    theme = row['theme']
    concept = row['concepts']

    # Tema ve kavram düğümleri ekleme
    theme_color = theme_colors.get(theme, 'gray')
    G.add_node(concept, color=theme_color)  # Kavram düğümlerine tema rengi ekleme
    G.add_node(participant, color=participant_colors.get(participant, 'gray'))  # Katılımcı düğümlerine renk ekleme
    G.add_edge(participant, concept, color=theme_color)  # Katılımcıdan kavrama kenar ekleme

    # Dönem düğümleri ekleme ve renklendirme
    if period not in G:
        G.add_node(period, color='lightblue')  # Dönem düğümlerine renk ekleme
    G.add_edge(period, concept, color='gray')  # Dönemden kavrama kenar ekleme
    G.add_edge(participant, period, color='lightblue')  # Katılımcıdan döneme kenar ekleme

# Spring layout ile pozisyon belirleme
pos = nx.spring_layout(G, k=0.5, iterations=50, seed=42)  # k değeri, düğümler arasındaki minimum mesafeyi kontrol eder

# Düğüm derecelerini hesaplama (düğüm büyüklükleri için)
degree_dict = dict(G.degree(G.nodes()))
sizes = [v * 200 for v in degree_dict.values()]  # Düğüm boyutlarını derecelerle orantılı olarak ayarlama

# Düğüm renkleri
node_colors = [nx.get_node_attributes(G, 'color').get(node, 'gray') for node in G]

# Görselleştirme
plt.figure(figsize=(16, 12))
nx.draw(G, pos, node_color=node_colors, node_size=sizes, edge_color='gray', alpha=0.6)  # Düğümlerin opaklığı

# Küçük harfe çevrilen etiketleri çizme
labels = {node: node.lower() for node in G.nodes()}
nx.draw_networkx_labels(G, pos, labels=labels, font_size=8, font_color='black', alpha=1)  # Yazıların opaklığı

plt.title('Network Graph for All Participants with Node Sizes by Degree')
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd

# Veri setini yükleme
df = pd.read_excel("/content/data.xlsx")

# Temalara göre renkler tanımlama
theme_colors = {
    'social environment': 'green',
    'physical environment': 'blue',
    'environmental meaning': 'red'
}

# Katılımcı renklerini tanımlama
participant_colors = {
    'participant 1': 'purple',
    'participant 2': 'orange',
    'participant 3': 'cyan',
    'participant 4': 'pink',
    # Diğer katılımcılar için renkler ekleyin
}


# Tüm katılımcıları içeren ağ grafiği oluşturma
G = nx.Graph()

for _, row in df.iterrows():
    participant = row['participants']
    period = row['periods']
    theme = row['theme']
    concept = row['concepts']

    theme_color = theme_colors.get(theme, 'gray')
    G.add_node(concept, color=theme_color)
    G.add_node(participant, color=participant_colors.get(participant, 'gray'))  # Düğüm rengi burada tanımlandı
    G.add_edge(participant, concept, color=theme_color)

    if period not in G:
        G.add_node(period, color='lightblue')
    G.add_edge(period, concept, color='gray')
    G.add_edge(participant, period, color='lightblue')

# Betweenness Centrality hesaplama
betweenness = nx.betweenness_centrality(G)
print("Betweenness Centrality:")
for node, centrality in betweenness.items():
    print(f"{node}: {centrality:.4f}")

# Degree Centrality hesaplama
degree_centrality = nx.degree_centrality(G)
print("\nDegree Centrality:")
for node, centrality in degree_centrality.items():
    print(f"{node}: {centrality:.4f}")

# Closeness Centrality hesaplama
closeness_centrality = nx.closeness_centrality(G)
print("\nCloseness Centrality:")
for node, centrality in closeness_centrality.items():
    print(f"{node}: {centrality:.4f}")

# Betweenness'e göre düğüm büyüklüklerini ayarlama
node_sizes = [v * 2000 for v in betweenness.values()]

# Düğüm renklerini alma
node_colors = [nx.get_node_attributes(G, 'color').get(node, 'gray') for node in G]

# Ağ grafiğini çizme
plt.figure(figsize=(14, 12))
pos = nx.spring_layout(G, k=0.5, seed=42)
nx.draw(G, pos, node_color=node_colors, node_size=node_sizes, edge_color='gray', alpha=0.9)

# Etiketleri çizme
labels = {node: node.lower() for node in G.nodes()}
nx.draw_networkx_labels(G, pos, labels=labels, font_size=8, font_color='black', alpha=0.6)

plt.title('Network Graph with Betweenness Centrality')
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
import numpy as np

# Veriyi yükleme ve ağ grafiğini oluşturma
df = pd.read_excel("/content/data.xlsx")

# Grafiği oluşturma
G = nx.Graph()

# Veriyi ağ grafiğine ekleme
for _, row in df.iterrows():
    participant = row['participants']
    period = row['periods']
    theme = row['theme']
    concept = row['concepts']

    G.add_node(participant)
    G.add_node(period)
    G.add_node(concept)

    G.add_edge(participant, concept)
    G.add_edge(period, concept)
    G.add_edge(participant, period)

# Betweenness Centrality hesaplama
betweenness = nx.betweenness_centrality(G)

# Degree Centrality hesaplama
degree_centrality = nx.degree_centrality(G)

# Closeness Centrality hesaplama
closeness_centrality = nx.closeness_centrality(G)

# Merkeziyet değerlerini bir DataFrame'e dönüştürme
centrality_df = pd.DataFrame({
    'Betweenness': betweenness,
    'Degree': degree_centrality,
    'Closeness': closeness_centrality
})

# Merkeziyet değerlerini karşılaştırmalı olarak çubuk grafiklerde gösterme
centrality_df.plot(kind='bar', figsize=(25, 8))
plt.title('Comparative Graph of Centrality Values')
plt.xlabel('Nodes')
plt.ylabel('Centrality Values')
plt.xticks(rotation=90)
plt.show()
```
